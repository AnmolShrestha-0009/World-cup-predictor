```python
import pandas as pd
import numpy as np
```

```python
import joblib
rf=joblib.load('rf_model.pkl')
xbg=joblib.load('xgb_model.pkl')
```

```python
def generate_rf_features(home_team, away_team):

    home = team_stats[home_team]
    away = team_stats[away_team]

    features = {
        "home_team_ranking": home["rank"],
        "away_team_ranking": away["rank"],
        "rank_diff": home["rank"] - away["rank"],

        "home_team_goals_scored_ratio": home["goals_scored_ratio"],
        "away_team_goals_scored_ratio": away["goals_scored_ratio"],

        "home_team_goals_conceded_ratio": home["goals_conceded_ratio"],
        "away_team_goals_conceded_ratio": away["goals_conceded_ratio"],
    }

    return pd.DataFrame([features])
```

```python
team_stats = {
    "France":      {"rank": 3,  "goals_scored_ratio": 2.321429, "goals_conceded_ratio":0.910714},
    "Morocco":     {"rank": 7, "goals_scored_ratio": 2.125000, "goals_conceded_ratio":0.575000},
    "Spain":       {"rank": 2,  "goals_scored_ratio": 2.491228, "goals_conceded_ratio":0.824561},
    "Belgium":     {"rank": 9,  "goals_scored_ratio": 2.260000, "goals_conceded_ratio":1.040000},
    "Norway":      {"rank": 31, "goals_scored_ratio": 2.594595	, "goals_conceded_ratio":1.216216},
    "England":     {"rank": 4,  "goals_scored_ratio": 2.537037, "goals_conceded_ratio":0.666667},
    "Argentina":   {"rank": 1,  "goals_scored_ratio": 1.909091, "goals_conceded_ratio":0.600000},
    "Switzerland": {"rank": 19, "goals_scored_ratio": 1.816327, "goals_conceded_ratio":1.265306}
}
```

```python
X = generate_rf_features("France", "Argentina")

X = X[rf.feature_names_in_]

prob = rf.predict_proba(X)
```

```python

def generate_xgb_features(home_team, away_team, neutral=1):
    home = team_stats[home_team]
    away = team_stats[away_team]

    features = {
        "neutral": neutral,

        "home_team_ranking": home["rank"],
        "away_team_ranking": away["rank"],

        "rank_diff": home["rank"] - away["rank"],

        "home_team_goals_scored_ratio": home["goals_scored_ratio"],
        "home_team_goals_conceded_ratio": home["goals_conceded_ratio"],

        "away_team_goals_scored_ratio": away["goals_scored_ratio"],
        "away_team_goals_conceded_ratio": away["goals_conceded_ratio"],
    }

    return pd.DataFrame([features])
```

```python
import pandas as pd

def predict(home_team, away_team, neutral=1):


    X_rf = generate_rf_features(home_team, away_team)
    X_xgb = generate_xgb_features(home_team, away_team, neutral)

    X_rf = X_rf[rf.feature_names_in_]
    X_xgb = X_xgb[xbg.feature_names_in_]


    rf_prob = rf.predict_proba(X_rf)[0][1]
    xgb_prob = xbg.predict_proba(X_xgb)[0][1]


    final_prob = (rf_prob + xgb_prob) / 2

    print(f"Match: {home_team} vs {away_team}")
    print("-" * 35)
    print(f"Random Forest : {rf_prob:.2%}")
    print(f"XGBoost       : {xgb_prob:.2%}")
    print(f"Final         : {final_prob:.2%}")

    if final_prob >= 0.5:
        print(f"\n🏆 Predicted Winner: {home_team}")
    else:
        print(f"\n🏆 Predicted Winner: {away_team}")

    return {
        "rf_probability": rf_prob,
        "xgb_probability": xgb_prob,
        "final_probability": final_prob
    }
```
